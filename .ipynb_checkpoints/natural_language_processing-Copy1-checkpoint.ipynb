# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from newsapi import NewsApiClient
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
from dotenv import load_dotenv
load_dotenv()

from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

%matplotlib inline
```

```python
# Read your api key environment variable
api_key = os.getenv("NEWS_API")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
bitcoin_news = newsapi.get_everything(q="Bitcoin", language="en", sort_by="relevancy", page_size=100)
bitcoin_news
```

```python
# Fetch the Ethereum news articles
ethereum_news = newsapi.get_everything(q="Ethereum", language="en", sort_by="relevancy", page_size=100)
ethereum_news
```

```python
#Print total articles
print(f"Total articles about Bitcoin: {bitcoin_news['totalResults']}")
print(f"Total articles about Ethereum: {ethereum_news['totalResults']}")
```

```python
#Show sample articles
print(bitcoin_news["articles"][0])
print()
print(ethereum_news['articles'][0])
```

```python
#Show sample sentment scores
article_=bitcoin_news["articles"][1]
text_ = article_["content"]
date_ = article_["publishedAt"][:10]
sentiment_ = analyzer.polarity_scores(text_)
sentiment_
```

```python
# Create the Bitcoin sentiment scores DataFrame

bitcoin_sentiments = []

for article in bitcoin_news['articles']:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        bitcoin_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral" : neu
        })
    except AttributeError:
        pass
    
# Create DataFrame
bitcoin_df = pd.DataFrame(bitcoin_sentiments)

# Reorder DataFrame columns
cols = ["date", "compound", "negative", "neutral", "positive","text"]
bitcoin_df = bitcoin_df[cols]

bitcoin_df           
```

```python
# Create the Ethereum sentiment scores DataFrame

ethereum_sentiments = []

for article2 in ethereum_news['articles']:
    try:
        text2 = article2["content"]
        date2 = article2["publishedAt"][:10]
        sentiment2 = analyzer.polarity_scores(text2)
        compound2 = sentiment2["compound"]
        pos2 = sentiment2["pos"]
        neu2 = sentiment2["neu"]
        neg2 = sentiment2["neg"]
        
        ethereum_sentiments.append({
            "text": text2,
            "date": date2,
            "compound": compound2,
            "positive": pos2,
            "negative": neg2,
            "neutral" : neu2
        })
    except AttributeError:
        pass
    
# Create DataFrame
ethereum_df = pd.DataFrame(ethereum_sentiments)

# Reorder DataFrame columns
cols = ["date", "compound", "negative", "neutral", "positive","text"]
ethereum_df = ethereum_df[cols]

ethereum_df
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# NLTK list of stopwords
print(stopwords.words('english'))

sw = set(stopwords.words('english'))
sw
```

```python
# Expand the default stopwords list if necessary
sw_addons = {'', 'chars', 'file', 'photo', 'representation', 'char', 'char ', 'said', 'sent', 'found', 'including', 'today', 'announced', 'week', 'basically', 'also'}
sw_added = set(sw.union(sw_addons))
sw_added
```

```python
print(len(sw_added))
print(len(sw))
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Create a list of the words
    words = word_tokenize(text)
    
    # Convert the words to lowercase
    lower_words = [word.lower() for word in words]
    
    # Remove the punctuation
     # Substitute everything that is NOT a letter with empty string
    regex = re.compile("[^a-zA-Z]")
    clean_words = [regex.sub('', lower_word) for lower_word in lower_words]

   
    # Remove the stop words
    cleaner_words = [clean_word for clean_word in clean_words 
                     if clean_word not in sw_added]
    
    # Lemmatize Words into root words
    tokens = [lemmatizer.lemmatize(cleaner_word) for cleaner_word in cleaner_words]
    
    return tokens
```

```python
# Create a new tokens column for bitcoin
# tokens = [tokenizer(text) for text in bitcoin_df['text']]
bitcoin_df['tokens'] = [tokenizer(text) for text in bitcoin_df['text']]
bitcoin_df
```

```python
# Create a new tokens column for ethereum
# tokens2 = [tokenizer(text2) for text2 in ethereum_df['text']]
ethereum_df['tokens'] = [tokenizer(text2) for text2 in ethereum_df['text']]
ethereum_df
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# concatenating the strings in the "text" column
btc_tokens =tokenizer(bitcoin_df["text"].str.cat())
```

```python
# Generate the Bitcoin N-grams where N=2
btc_bigrams = ngrams(btc_tokens, n=2)
btc_bigrams
```

```python
# Frequency of bitcoin_ngrams(N=2)
btc_bigram_counts = Counter(btc_bigrams).most_common(20)
btc_bigram_counts
```

```python
# Generate the Ethereum N-grams where N=2
eth_bigrams = ngrams(tokenizer(ethereum_df["text"].str.cat()), n=2)
eth_bigrams
```

```python
# Frequency of ethereum_ngrams(N=2)
eth_bigram_counts = Counter(eth_bigrams).most_common(20)
eth_bigram_counts
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
bitcoin_tokens = tokenizer(bitcoin_df["text"].str.cat())
token_count(bitcoin_tokens, N=10)
```

```python
btc_top_10 = dict(token_count(bitcoin_tokens, N=10))
pd.DataFrame(list(btc_top_10.items()), columns=['word', 'count'])
```

```python
# Get the top 10 words for Ethereum
ethereum_tokens = tokenizer(ethereum_df["text"].str.cat())
token_count(ethereum_tokens, N=10)
```

```python
eth_top_10 = dict(token_count(ethereum_tokens, N=10))
pd.DataFrame(list(eth_top_10.items()), columns=['word', 'count'])
```

# Word Clouds

In this section, you will generate word clouds for each coin to summarize the news for each coin

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = [20.0, 10.0]
```

```python
# function to generate wordcloud
def make_wordcloud(text, name):
    df_cloud = WordCloud(width=500, colormap="viridis").generate(text)
    plt.imshow(df_cloud)
    plt.axis('off')
    font_dict={'fontsize': 50, 'fontweight': "bold"}    
    plt.title(f"{name} News Word Cloud", fontdict = font_dict)    
```

```python
# Process to generate ethereum news word cloud
bitcoin_df["tokens_2"] = [" ".join(x) for x in bitcoin_df.tokens]
bitcoin_df.head(2)
```

```python
# generate bitcoin news word cloud   
make_wordcloud(bitcoin_df.tokens_2.str.cat(), "Bitcoin")  
```

```python
# Process to generate ethereum news word cloud
ethereum_df["tokens_2"] = [" ".join(x) for x in ethereum_df.tokens]
ethereum_df.head(2)
```

```python
# generate ethereum news word cloud
make_wordcloud(ethereum_df.tokens_2.str.cat(), "Ethereum") 
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
#!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# sanity check
bitcoin_df.head(2)
```

```python
# Concatenate all of the bitcoin text together
bitcoin_text = bitcoin_df.tokens_2.str.cat()
bitcoin_text
```

```python
# Run the NER processor on all of the text
bitcoin_doc = nlp(bitcoin_text)

# Add a title to the document
bitcoin_doc.user_data["title"] = "Bitcoin Named Entity Recognition"
```

```python
# Render the visualization
displacy.render(bitcoin_doc, style='ent')
```

```python
# List all Entities1
bitcoin_entities = [[btc_entity.text, btc_entity.label_] for btc_entity in bitcoin_doc.ents]
bitcoin_entities
```

```python
# List all Entities2
for btc_entity in bitcoin_entities:
    print(btc_entity[0], btc_entity[1])
```

```python
# create dataframe of Entities
bitcoin_entities2 = [{"text":btc_entity.text, "label": btc_entity.label_} for btc_entity in bitcoin_doc.ents]
pd.DataFrame(bitcoin_entities2)
```
